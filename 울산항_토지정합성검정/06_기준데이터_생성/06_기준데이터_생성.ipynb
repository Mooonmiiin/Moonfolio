```python
import pandas as pd
import numpy as np

# 파일 불러오기
df = pd.read_excel("토지이음_수집데이터.xlsx")

# -------------------------
# 문자열 정리
# -------------------------
def clean_text(x):
    if pd.isna(x):
        return ""
    return str(x).strip()

df["소재지"] = df["소재지"].apply(clean_text)

# -------------------------
# 면적 열 숫자형 변환
# -------------------------
def to_float(x):
    if pd.isna(x):
        return np.nan
    s = str(x).replace(",", "").strip()
    if s == "":
        return np.nan
    try:
        return float(s)
    except:
        return np.nan

df["면적"] = df["면적"].apply(to_float)

# -------------------------
# 1. 지목 수정
# -------------------------
jimok_fix = {
    "울산광역시 남구 매암동 139-43": "창고용지",
    "울산광역시 남구 매암동 472-50": "창고용지",
    "울산광역시 남구 황성동 863-1": "도로",
    "울산광역시 남구 황성동 864-1": "도로",
}

for addr, value in jimok_fix.items():
    df.loc[df["소재지"] == addr, "지목"] = value

# -------------------------
# 2. 면적 수정
# -------------------------
area_fix = {
    "울산광역시 남구 매암동 1-10": 884,
    "울산광역시 남구 매암동 139-82": 2519,
    "울산광역시 남구 매암동 472-42": 1800,
    "울산광역시 남구 여천동 187-10": 49472,
    "울산광역시 남구 여천동 187-12": 107564,
}

for addr, value in area_fix.items():
    df.loc[df["소재지"] == addr, "면적"] = float(value)

# -------------------------
# 저장
# -------------------------
df.to_excel("기준데이터.xlsx", index=False)

print("수정 완료")
```

```python
import pandas as pd

# =========================
# 1. 파일 불러오기
# =========================
df_eum = pd.read_excel("기준데이터.xlsx")
df_target = pd.read_excel("비교대상데이터.xlsx")


# =========================
# 2. 소재지 정리 (핵심🔥)
# =========================
def clean_text(x):
    if pd.isna(x):
        return ""
    return " ".join(str(x).strip().split())

df_eum["소재지"] = df_eum["소재지"].apply(clean_text)
df_target["소재지"] = df_target["소재지"].apply(clean_text)


# =========================
# 3. 병합 (토지이음 기준)
# =========================
merged = df_eum.merge(
    df_target,
    on="소재지",
    how="left"
)


# =========================
# 4. 컬럼 순서 정리
# =========================
merged = merged[[
    "소재지",
    "지목",
    "면적",
    "토지이동사유",
    "토지이동일",
    "상태",
    "고시면적",
    "토지면적",
    "출자면적",
    "고시지목",
    "토지지목",
    "출자지목"
]]


# =========================
# 5. 저장
# =========================
merged.to_excel("통합데이터(기준).xlsx", index=False)

print("완료")
print("행 개수:", len(merged))
```
