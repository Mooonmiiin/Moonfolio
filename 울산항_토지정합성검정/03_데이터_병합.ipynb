```python
import pandas as pd

# =========================
# 1. 파일 불러오기
# =========================
df_eum = pd.read_excel("토지이음_수집데이터.xlsx")
df_target = pd.read_excel("비교대상데이터.xlsx")


# =========================
# 2. 소재지 정리 (핵심🔥)
# =========================
def clean_text(x):
    if pd.isna(x):
        return ""
    return " ".join(str(x).strip().split())

df_eum["소재지"] = df_eum["소재지"].apply(clean_text)
df_target["소재지"] = df_target["소재지"].apply(clean_text)


# =========================
# 3. 병합 (토지이음 기준)
# =========================
merged = df_eum.merge(
    df_target,
    on="소재지",
    how="left"
)


# =========================
# 4. 컬럼 순서 정리
# =========================
merged = merged[[
    "소재지",
    "지목",
    "면적",
    "토지이동사유",
    "토지이동일",
    "상태",
    "고시면적",
    "토지면적",
    "출자면적",
    "고시지목",
    "토지지목",
    "출자지목"
]]


# =========================
# 5. 저장
# =========================
merged.to_excel("통합데이터.xlsx", index=False)

print("완료")
print("행 개수:", len(merged))
```
